#### 3.1.1 What is the revenue and profit generated each year?<a id="part3.1.1"></a>

```python
df['Revenue'] = df['UnitPrice in USD'] * df['Quantity'] * (1 - df['Unit Discount in %']) 
df['Profits'] = df['Revenue'] - (df['UnitCost in USD'] * df['Quantity'])
```

```python
# Retrieve data to create the plot
revenue1 = df.groupby('Order Year')['Revenue'].sum().round(decimals = 2)
profits1 = df.groupby('Order Year')['Profits'].sum().round(decimals = 2)
year = revenue1.index

# Set the label locations and width of the bars
x = np.arange(len(year))
width = 0.35

# Create a figure and subplot for revenue chart
fig, ax = plt.subplots(1, 1, figsize = (10, 6))

# Plot side-by-side bars
bar1 = ax.bar(x - width/2, revenue1, width, label = 'Revenue')
bar2 = ax.bar(x + width/2, profits1, width, label = 'Profits')

# Customise the plot
ax.set_title('Revenue And Profits From the Past 5 Years', fontsize = 20)
ax.set_xlabel('Year', fontsize = 15)
ax.set_ylabel('Amount ($)', fontsize = 15)
ax.set_xticks(x)
ax.set_xticklabels(year)
ax.set_xlim(left = -1, right = 5)
ax.legend()

# Create a function to display data labels/annotations for each bar
def autolabel(bars):
    for each in bars:
        height = each.get_height()
        ax.annotate('{}'.format(height),
                    xy = (each.get_x() + each.get_width() / 2, height),
                    xytext = (0, 2),
                    textcoords = 'offset points',
                    ha = 'center', va = 'bottom')

# Use function to display data labels for both bars
autolabel(bar1)
autolabel(bar2)

# Set tight layout for figure
fig.tight_layout()

# Display plot
plt.show()

# Save plot as 'mslegend.png' in same folder as notebook
#plt.savefig('mslegend.png', dpi = 300)
```

This grouped bar graph shows the revenue and profits made by DART from 2014 to 2018. The <font color = 'blue'>blue</font> bars show the Revenue earned each year while the <font color = 'orange'>orange</font> bars show the Profit made each year.

As sales were only generated from quarter 4 of 2014 onwards, there isnt much to see in terms of revenue and profit generated for the year of 2014. 

2016 saw an increase of almost 80% in Revenue while the profit almost doubled.

2017 saw another increase of more than 100% in terms of Revenue and profit.

2018, however, saw a huge fall in revenue and profit made by almost 60%.

This needs to be improved. 1 way of doing so is improving our Advertising by investing more into it.

#### 3.1.2 Which quarter is the most productive in terms of sales?<a id="part3.1.2"></a>

```python
quarters = ['Qtr 1', 'Qtr 2', 'Qtr 3', 'Qtr 4']
df['Order Quarter'] = pd.Categorical(df['Order Quarter'], categories = quarters, ordered = True )

#Year 2014
year2014 = df[df['Order Year'] == 2014]
sales2014 = year2014.groupby('Order Quarter')['OrderID'].count()
quarter2014 = sales2014.index
#Year 2015
year2015 = df[df['Order Year'] == 2015]
sales2015 = year2015.groupby('Order Quarter')['OrderID'].count()
quarter2015 = sales2015.index
#Year 2016
year2016 = df[df['Order Year'] == 2016]
sales2016 = year2016.groupby('Order Quarter')['OrderID'].count()
quarter2016 = sales2016.index
#Year 2017
year2017 = df[df['Order Year'] == 2017]
sales2017 = year2017.groupby('Order Quarter')['OrderID'].count()
quarter2017 = sales2017.index
#Year 2018
year2018 = df[df['Order Year'] == 2018]
sales2018 = year2018.groupby('Order Quarter')['OrderID'].count()
quarter2018 = sales2018.index

#Create a figure and subplot for revenue chart
fig, ax = plt.subplots(1, 1, figsize=(15,5))

#Plot bar
ax.plot(quarters, sales2014, color = 'c', label = '2014')
ax.plot(quarters, sales2015, color = 'r', label = '2015')
ax.plot(quarters, sales2016, color = 'y', label = '2016')
ax.plot(quarters, sales2017, color = 'g', label = '2017')
ax.plot(quarters, sales2018, color = 'b', label = '2018')

#Customise the plot
ax.set_title('Number of sales in each quarter for each year', fontsize = 20)
ax.set_xlabel('Quarters', fontsize = 15)
ax.set_ylabel('Number of Sales', fontsize = 15)
    
#turn grid on
ax.grid(alpha = 0.2, linestyle = '--')
ax.legend()
# Set tight layout for figure
fig.tight_layout()

# Display plot
plt.show()
```

This bar chart shows the amount of sales made in each quarter for each year.

As you can see, Quarter 4 has generated the most amount of sales for every single year apart from 2018, making it the most productive quarter in terms of sales.

Quarters 2 and 3 are the least productive but somewhat similar figures, with Quarter 2 generating lesser than Quarter 3, meaning that the number of sales made is very low but consistent between the months of April to September, but remain lower than the other 6 months.

This needs to be improved upon also.

#### 3.1.3 What is the revenue generated by each country?<a id="part3.1.3"></a>

```python
#retrieve data to create the plot
revenue2 = df.groupby('Customer Country')['Revenue'].sum().round(decimals = 2)
country1 = revenue2.index

#Create a figure and subplot for revenue chart
fig, ax = plt.subplots(1, 1, figsize=(15,6))

#Plot bar
ax.bar(country1, revenue2, label = 'Revenue Generated')

#Customise the plot
ax.set_title('Revenue Generated by Each Country', fontsize = 20)
ax.set_xlabel('Countries', fontsize = 15)
ax.set_ylabel('Revenue', fontsize = 15)

#label the points
for x, y in zip(country1, revenue2): #zip joins x and y coordinate in pairs
    label = y #value of data label to display
    plt.annotate(label, # this is the data label to display
                 (x,y), # point to insert data label
                 textcoords="offset points", # how to position the text
                 xytext=(0,0), # distance from text to points (x,y)
                 ha='center', size = 15) # horizontal alignment can be left, right or center

#turn grid on
ax.grid(alpha = 0.2, linestyle = '--')

# Set tight layout for figure
fig.tight_layout()

```

#### 3.1.5 Which product category generated the most revenue and profit?<a id="part3.1.5"></a>

```python
# Define x-axis and y-axis data values
product_category = ['Baby Clothes', 'Childrens Clothes', 'Mens Clothes', 'Mens Footwear', 'Sportswear', 'Swimwear', 'Womens Clothes', 'Womens Footwear' ]
revenue = df.groupby('Product CategoryName')['Revenue'].sum().round(decimals = 2)
profits = df.groupby('Product CategoryName')['Profits'].sum().round(decimals = 2)

# Create a figure and subplot for revenue chart
fig, ax_r = plt.subplots(1,1,figsize=(15,8))

# Plot revenue chart
ax_r.bar(product_category, revenue, width=0.35, label='Revenue')
ax_r.set_ylim(bottom = 0, top = 600000)
ax_r.set_title('Revenue and Profits Generated by each Product Category', fontsize = 20)
ax_r.set_ylabel('Revenue', fontsize = 15) # first y-axis (on the left)

# Instantiate a second axes that shares the same x-axis
ax_p = ax_r.twinx()

# Plot profits chart
ax_p.plot(product_category, profits, 'g', label = 'Profits')
ax_p.set_ylim(bottom = 0, top = 600000)
ax_p.set_ylabel('Profits', fontsize = 15) # second y-axis (on the right)

# Set location of legend for both axes
fig.legend(loc=(0.74,0.78))

# Set tight layout for figure
fig.tight_layout()

# Display plot
plt.show()
```

#### 3.1.6 What is the distribution for each product category in terms of revenue?<a id="part3.1.6"></a>

```python
# Define x-axis and y-axis data values
product_category = df['Product CategoryName'].unique()

#Create an empty list
prod_cat = []

#Generate distribution of profits for each product category
for i in product_category:
    temp = df[df["Product CategoryName"]== i]
    temp = temp['Revenue'].tolist()
    prod_cat.append(temp)

# Create a figure and subplot for profits chart
fig, ax = plt.subplots(1,1,figsize=(15,8))

# Create Multiple Box Plots
plt.boxplot(prod_cat, labels = product_category)

#Customise the plot
ax.set_title('Distribution of Revenue for each product category', fontsize = 20)
ax.set_xlabel('Product Category Name', fontsize = 15)
ax.set_ylabel('Revenue', fontsize = 15)
ax.set_ylim(bottom = -1000, top = 5000)

#Display plot
plt.show()
```

This visual is a box and whisker plot that compares the revenue made by each product category from 2014 to 2018. 

It can be inferred that Men's Footwear generally generates the most amount of revenue as it has the highest median revenue. However, it also has the largest Inter-Quartile Range(IQR), allowing for a larger chance if the values being inconsistent. 

It can also be inferred from the viusal that Swimwear generally generates the least amount of revenue as it has the lowest median revenue. However, it also has the smallest IQR, making the values more consistent.

As there are way to many outliers in the data,it is unclear whether the average and median of the Revenue can be an accurate representation of the revenue generated by each Product Category.
